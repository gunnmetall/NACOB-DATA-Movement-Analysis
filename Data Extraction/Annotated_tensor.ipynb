```python
sync=pd.read_excel("/content/Sync File.xlsx")
Participant_tensor = np.empty((0, 4, 14400, 5))
tensor_participant=np.empty((0,14400,5))
dir=os.getcwd()
columns_for_square = ['Gyr_X','Gyr_Y','Gyr_Z']
columns_to_read=['PacketCounter','Gyr_X','Gyr_Y','Gyr_Z']
video=[]
person=[]
folder=[]
Shank_df=np.zeros([14400,5])
Trunk_df=np.zeros([14400,5])
Foot_df=np.zeros([14400,5])
Thigh_df=np.zeros([14400,5])
check=np.zeros([176,5])
count=-1
for x in path:
  dir1=dir+"/"+x
  person.append(dir1)
  count=count+1
  for filename in os.listdir(dir1):
      f = os.path.join(dir1, filename)
      if os.path.isfile(f):
        video.append(f)
      else:
        print(f)
        check[count][0]=count
        start_row=int(sync.iloc[count][9])
        if start_row>0:
          continue
        start_row=abs(start_row)
        end_row=start_row+14400
        for filename2 in os.listdir(f):
          f2=os.path.join(f, filename2)
          file_size = os.path.getsize(f2)
          if(file_size>300):
              if filename2.find("Shank")!=-1:
                shank_df=pd.read_csv(f2,usecols=columns_to_read,sep=',',skiprows=[0])
                shank_df = shank_df.applymap(lambda x: x if isinstance(x, (int, float)) else 0)
                shank_df['Magnitude'] = np.sqrt(np.sum(shank_df[columns_for_square] ** 2, axis=1))
                shank_df = shank_df.iloc[start_row:end_row]
                shank_arr=shank_df.values
                if(shank_arr.shape==(14400,5)):
                  Shank_df = np.vstack((Shank_df, shank_arr))
                  check[count][1]=1
              elif filename2.find("Trunk")!=-1:
                trunk_df=pd.read_csv(f2,usecols=columns_to_read,sep=',',skiprows=[0])
                trunk_df = trunk_df.applymap(lambda x: x if isinstance(x, (int, float)) else 0)
                trunk_df['Magnitude'] = np.sqrt(np.sum(trunk_df[columns_for_square] ** 2, axis=1))
                trunk_df = trunk_df.iloc[start_row:end_row]
                trunk_arr=trunk_df.values
                if(trunk_arr.shape==(14400,5)):
                  Trunk_df = np.vstack((Trunk_df, trunk_arr))
                  check[count][2]=1
              elif filename2.find("Foot")!=-1:
                foot_df=pd.read_csv(f2,usecols=columns_to_read,sep=',',skiprows=[0])
                foot_df = foot_df.applymap(lambda x: x if isinstance(x, (int, float)) else 0)
                foot_df['Magnitude'] = np.sqrt(np.sum(foot_df[columns_for_square] ** 2, axis=1))
                foot_df = foot_df.iloc[start_row:end_row]
                foot_arr=foot_df.values
                if(foot_arr.shape==(14400,5)):
                  Foot_df = np.vstack((Foot_df, foot_arr))
                  check[count][3]=1
              elif filename2.find("Thigh")!=-1:
                thigh_df=pd.read_csv(f2,usecols=columns_to_read,sep=',',skiprows=[0])
                thigh_df = thigh_df.applymap(lambda x: x if isinstance(x, (int, float)) else 0)
                thigh_df['Magnitude'] = np.sqrt(np.sum(thigh_df[columns_for_square] ** 2, axis=1))
                thigh_df = thigh_df.iloc[start_row:end_row]
                thigh_arr=thigh_df.values
                if(thigh_arr.shape==(14400,5)):
                  Thigh_df = np.vstack((Thigh_df, thigh_arr))
                  check[count][4]=1
              elif check[count][1]==1 and check[count][2]==1 and check[count][3] and check[count][4]==1:
                tensor_participant = np.stack((shank_arr,trunk_arr, foot_arr, thigh_arr))
              else:
                print("exception found at this file :",filename2)

          else:
            pass

          folder.append(f2)
          print(filename2)
  if(tensor_participant.shape==(4,14400,5)):
    Participant_tensor = np.concatenate((Participant_tensor, tensor_participant[np.newaxis]), axis=0)
```
