```python
import pickle
import pandas  as pd
from datetime import date
from datetime import time
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
df_xdot = pd.read_excel('Ground Truth Sync.xlsx',skiprows=[0])
df_xdot["Hours"].fillna("NULL", inplace = True)
df_xdot
```

```python
count=0
ss="P_0"
i=0
Good_one=[]
Bad_one=[]
sss=[]
print("Serial Number   Timestamp_xdot         Participant")
while(i<176):
    ss="P_0"
    if(df_xdot.loc[i]['Hours']=="NULL"):
        Bad_one.append(i+1)
        i=i+1
        continue
    else:
        if(i+1 < 10):
            ss=ss+"00"+str(i+1)
        elif(i+1 < 100):
            ss=ss+"0"+str(i+1)
        else:
            ss=ss+str(i+1)
        sss.append(ss)
        s=str(df_xdot.loc[i]['Date'])
        s=s.replace('00:00:00','')
        s=s+str(int(df_xdot.loc[i]['Hours']))+":"+str(int(df_xdot.loc[i]['Minutes']))+":"+str(int((df_xdot.loc[i]['Seconds'])))
        s=s.replace('.','')
        Good_one.append(i+1)
        count+=1
        print(count,"         ",s,'     ',ss)
        i=i+1
```

```python
#Number of participants we have after subtracting the bad videos
print(count)
```

```python
#these are the Participants that we are unable to sync because either the video was corrupted or there was no server clock while screen recording.
print(Bad_one)
```
